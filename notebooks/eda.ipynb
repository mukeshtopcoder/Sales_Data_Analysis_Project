```python
import pandas as pd
```

```python
df = pd.read_excel(r"../dataset/Financial_Sample.xlsx")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())
```
